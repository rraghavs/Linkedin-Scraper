```python
# Company Lists 
Company_name=[]
Company_follower=[]
Company_no_of_emp=[]
Company_industry=[]


# Details Lists
Job_involment=[]
Job_level=[]
Job_total_application=[]


driver = webdriver.Chrome()
driver.get("https://www.linkedin.com/")
driver.maximize_window()
sleep(2)



username_field = driver.find_element(By.ID,"session_key")
username_field.send_keys('teja.polipogu@gmail.com')
    
    
password_field = driver.find_element(By.ID,"session_password")
password_field.send_keys('9182284060')


next_b_b = driver.find_element(By.XPATH, '//*[@id="main-content"]/section[1]/div/div/form/div[2]/button')
if next_b_b.is_displayed():
    next_b_b.click()
sleep(30)  


for i in Job_Link:
    driver.get(i)
    sleep(5)
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # Company Details 
    try:
        company_name=driver.find_element(By.XPATH,'//div[@class="artdeco-entity-lockup__title ember-view t-20"]/a')
        print(company_name.text)
        Company_name.append(company_name.text)
    except:
#         print("NA")
        Company_name.append("NA")
    
    try:
        company_follower=driver.find_element(By.XPATH,'//div[@class="artdeco-entity-lockup__subtitle ember-view t-16"]')
#         print(company_follower.text)
        Company_follower.append(company_follower.text)
    except:
        print("NA")
        Company_follower.append("NA")
    
    try:
        company_no_of_emp = driver.find_element(By.XPATH,'//div[@class="t-14 mt5"]/span[1]')
#         print(company_no_of_emp.text)
        Company_no_of_emp.append(company_no_of_emp.text)
    except:
#         print("NA")
        Company_no_of_emp.append("NA")
    
    try:
        company_industry = driver.find_element(By.XPATH,'//div[@class="t-14 mt5"]')
        print(company_industry.text)
        Company_industry.append(company_industry.text)
    except:
#         print("NA")
        Company_industry.append("NA")
       

# Job Details 
    try:
        job_involment = driver.find_element(By.XPATH,'//span[@class="ui-label ui-label--accent-3 text-body-small"]/span[1]')
#         print(job_involment.text)
        Job_involment.append(job_involment.text)
    except:
#         print("NA")
        Job_involment.append("NA")
    
    try:
        job_level = driver.find_element(By.XPATH,'//span[@class="job-details-jobs-unified-top-card__job-insight-view-model-secondary"][2]')
#         print(job_involment.text)
        Job_level.append(job_level.text)
    except:
#         print("NA")
        Job_level.append("NA")
    
    try:
        job_total_application = driver.find_element(By.XPATH,'//div[@class="job-details-jobs-unified-top-card__primary-description-without-tagline mb2"]/span[5]')
#         print(job_total_application.text)
        Job_total_application.append(job_total_application.text)
    except:
#         print("NA")
        Job_total_application.append("NA")
    

driver.quit()
```
